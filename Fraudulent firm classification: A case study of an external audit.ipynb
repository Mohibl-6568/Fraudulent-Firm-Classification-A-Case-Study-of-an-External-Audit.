```python
import numpy as np
import pandas as pd
from pyswarm import pso

X = df.drop(columns=["LOCATION_ID", "Risk"])
y = df["Risk"]

X = X.fillna(X.mean())

F = X.values  # Feature matrix (T x n)
alpha = y.values  # Target vector (T)

T, n = F.shape

def objective_function(w):
    error = 0
    for i in range(T):
        prediction = np.sum(w * F[i])
        error += (alpha[i] - prediction) ** 2
    return error

#Define bounds for each weight
lb = [-10] * n  # lower bound for each weight
ub = [10] * n   # upper bound for each weight

#Run PSO
best_weights, best_error = pso(objective_function, lb, ub, swarmsize=30, maxiter=100)

print("Best Weights found:", best_weights)
print("Best Error achieved:", best_error)
```

```python
# After running PSO and getting best_weights

# Step 1: Create a dataframe of features and their optimized weights
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Weight": best_weights,
    "Abs_Weight": np.abs(best_weights)
})

# Step 2: Sort by absolute weight (importance)
feature_importance_sorted = feature_importance.sort_values(by="Abs_Weight", ascending=False)

# Step 3: Pick the top 9 features
top_9_features = feature_importance_sorted.head(9)

print("Top 9 features based on PSO optimization:")
print(top_9_features[["Feature", "Weight"]])
```

```python
top_9_feature_names = top_9_features["Feature"].tolist()

df_final = df[top_9_feature_names + ['Risk']]  # Also keep 'Risk' as target

# Now df_top9 contains only the important features + the Risk column
print(df_final.head())
```
